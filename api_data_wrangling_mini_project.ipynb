1. Collect data from the Frankfurt Stock Exchange, for the ticker AFX_X, for the whole year 2017.

```python
year=requests.get('https://www.quandl.com/api/v3/datasets/FSE/AFX_X.json?api_key=7up7a3-xNp8abz4VJGsq&start_date=2017-01-01&end_date=2018-01-01')
```

2. Convert the returned JSON object into a Python dictionary

```python
data=year.json()['dataset']
```

3. Calculate what the highest and lowest opening prices were for the stock in this period

```python
year_data=data['data']
col_names=data['column_names']
print(col_names)
len(year_data)
```

```python
import pandas as pd
df=pd.DataFrame(year_data, columns=col_names)
df.head()
```

```python
highest_open=df['Open'].max()
lowest_open=df['Open'].min()
print(highest_open)
print(lowest_open)
```

The highest opening is 53.11 and the lowest opening is 34.0 for this period.

4. What was the largest change in one day?

```python
largest_change=(df['High']-df['Low']).max()
print(largest_change)
```

The largest change in one day is 2.81.

5. What was the largest change in any two days?

```python
changes_two = [df['Close'][i+1]-df['Close'][i] for i in range(len(year_data)-1)]
from numpy import array
largest_two_day_change= array(changes_two).max()
print(largest_two_day_change)
```

The largest change in any two days is 2.559999

6. What was the average trading?

```python
average_trading=df['Traded Volume'].sum()/len(data)
print(average_trading)
```

The average trading is 1082224.095

Optional 7. What was the median trading

```python
import numpy as np
median_trading=np.median(df['Traded Volume'])
print(median_trading)
```
